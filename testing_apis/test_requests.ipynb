setting up the credentials to use the reddit api

```python
base_url = 'https://www.reddit.com/'
data = {'grant_type': 'password',
        'username': reddit_config.REDDIT_USERNAME,
        'password': reddit_config.REDDIT_PASSWORD}
auth = requests.auth.HTTPBasicAuth(reddit_config.APP_ID,
                                   reddit_config.APP_SECRET)
r = requests.post(base_url + 'api/v1/access_token',
                  data=data,
                  headers={'user-agent': 'APP-NAME by REDDIT-USERNAME'},
                  auth=auth)
d = r.json()
```
