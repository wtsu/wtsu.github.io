```python
#Create a Dummy PD Series
import pandas as pd
data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=['a', 'b', 'c', 'd'])
data

# PD Series are like dictionaries. You can access an element by specifying the key
data['b']

# You can use dictionary related actions on a pd series
## is an index in a dictionary
'a' in data

##what are the keys
data.keys()

##list each key-pair combination
list(data.items())

##create new key-pairs in the dictionary
data['e'] = 1.25
data

#Series can be worked with like a numpy array
##select a particular range of indices
###when using the key value for slicing, the final element is included
data['a':'c']

##slicing the the index value of the indices
###when using the index value for slicing, the final element is excluded
data[0:3] 

##using a mask
data[(data > 0.3) & (data < 0.8)]

##looking at specific indices
data[['a','e']]

#If the key is a number, it can be confusing if you are slicing via the index or the key value
data = pd.Series(['a', 'b', 'c'], index=[1, 3, 5])
data

##this returns the value marked within the key = 1
data[1]

##this returns the value associated with the index value
data[1:3]

#reduce confusion by using loc, iloc

##.loc[] will always explicitly look for the key value
data.loc[1]
data.loc[1:3]

##.iloc[] will reference via the indice
data.iloc[1]
data.iloc[1:3]

#creating a dataframe
area = pd.Series({'California': 423967, 'Texas': 695662,
                  'New York': 141297, 'Florida': 170312,
                  'Illinois': 149995})
pop = pd.Series({'California': 38332521, 'Texas': 26448193,
                 'New York': 19651127, 'Florida': 19552860,
                 'Illinois': 12882135})
data = pd.DataFrame({'area':area, 'pop':pop})

# a dataframe is a dictionary where each key refers to a pd.Series object
##to access the pd.series related to a key
data['area']
##or equivalently
data.area
data['area'] is data.area
##this way of accessing data though will not work if the dataframe key is not a string or is the same as a reserved method

##This method can be used to create new values
data['density'] = data['pop']/data['area']
data

## a pd dataframe can also be viewed as a 2d array
data.values

##we can use functions like transpose
data.T

##you cannot treat a pd dataframe as a 2d numpy array perfectly
##an index returns a row in Numpy
type(data.values)
data.values[0]
##an index returns a col in a DF
data['area']

##for array style indexing, use loc,iloc,ix

#returns the first 3 rows, and the first 2 colums
data.iloc[:3,:2]

#returns all the rows up to New York, and all the colums up to pop
data.loc[:'New York',:'pop']

#ix is a hybrid approach. It allos for both slicing by index value and key value
#first three rows, and the pop column
#ix[] is being deprecated however. ix is also subject to confusion if the key value is an integer
data.ix[:3,:'pop']

#using a mapping and column selection
##get the rows where density is >100 and get the columns 'pop','density'
data.loc[data.density>100,['pop','density']]

##you can specivy the particular cell to modify too
##change the value in row index = 0 and column index = 2 to be 90
data.iloc[0,2] = 90
data

#indexing refers to columns, slicing refers to rows
#this is a slice b/c it gets the rows between florida and illinois
#this differs from indexing b/c you are noting a range of values
#indexing would bet data['column_name']
#slicing would be data['row_index;:'row_index']
data['Florida':'Illinois']
data[1:3]

#masking is also done on a row level
#which rows have density>100
data[data.density>100]

```
