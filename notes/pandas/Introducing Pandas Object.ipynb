```python
import pandas as pd
import numpy as np
print(np.__version__)
print(pd.__version__)
```

```python
data = pd.Series([0.25,0.5,0.75,1.0])
data.values #this is the np array
data.index
data[1]
data[1:3]

#pd.Series is like special type of dictionary. It maps a value in 
#and np array with an index

#index can be explicitly declared
data = pd.Series([0.25,0.5,0.75,1.0],
                index = ['a','b','c','d'])
data['b']

#indices don't have to be in order
data = pd.Series([0.25,0.5,0.75,1.0],
                index = [2,5,3,7])
data[3]

#pd.Series as a dictionary
population_dict = {'CA':1000,
                   'TX':3000,
                   'NY':7000,
                   'FL':1823,
                   'IL':1232}
pop = pd.Series(population_dict)
pop['TX'] #note how the index is auto created from the dictionary

#unlike a dictionary, pd.Series alllows for array style slicing
pop['CA':'FL']

#pd.Series(data, index=index)

pd.Series([2,4,6])

#the 'data' can be repeated if you introduce multiple indices
pd.Series(5, index = [100,200,300])

#the 'data' can be a dictionary. the indice is implied through the dictionary
pd.Series({2:'a',
           1:'b',
           3:'c'})

#you can use the index to limit the data values to what you want
pd.Series({2:'a', 1:'b', 3:'c'},
         index = [1,2])

area_dict = {'CA': 423967, 'TX': 695662, 'NY': 141297,
             'FL': 170312, 'IL': 149995}
area = pd.Series(area_dict)

#use the two pd.Series area and pop to create a dataframe
#this effectively does a full outer join where the key used is the pd.Series indices
states = pd.DataFrame({'area':area,
              'population':pop 
})
states.index
states.columns

#you can think of a Dataframe also as a dictionary.
#Each column is the key whose value is a series
states['area']

#Dataframe construction in multiple flavors
#DF via a single series
pd.DataFrame(pop,
            columns=['population'])

#DF with a bunch of dictionaries

#generate these dictionaries using a list comprehension
data = [{'a':i, 'b':2*i} for i in range(3)]
#for reference, the list comprehension above is a more concise way of doing the following
x = []
for i in range(3):
    x.append({'a':i, 'b':2*i})
x == data

pd.DataFrame(data)

#if the dictionaries used to construct the DF is missing keys, it will fill with na
a = {'a':1,'b':2}
b = {'a':2,'c':3}
pd.DataFrame([a,b])

#DF construction by using a bunch of series objects
pd.DataFrame({'population':pop,
             'area':area})

#DF construction from 2d numpy array
pd.DataFrame(np.random.rand(3,2),
            columns = ['foo','bar'],
            index = ['a','b','c'])

#NEED TO REVIEW
#DF construction through a structured numpy array
A = np.zeros(3, dtype=[('A','i8'),('B','f8')])
pd.DataFrame(A)

#Most DF have the index. This index is its own object
#construct an index object
ind = pd.Index([2,3,5,7,11])
ind

#NEED TO REVIEW array slicing
#like an array, we can use array slicing
ind[1]
ind[::2]

#the difference between an index and an array is that an index cannot be changed
#this index immutability is useful if you need to make sure you are using the same index across dataframes
try:
    ind[1] = 0
except:
    a = 'failure'
    #print(a)

#index object supports unions, intersections, differences, and other combos
indA = pd.Index([1,2,3,4])
indB = pd.Index([2,3,4,5,6])

#what indices exist in both
indA & indB

#what is the union of indicies
indA | indB

#what doesn't exist in both
indA ^ indB
```
