```python
import numpy as np
np.array(range(5))

#unlike lists, np.array will only contain one datatype
#it will therefore convert the int => decimals
np.array([1,2,3,4,5.0])
```

```python
#to explicitly set datatype, us the "dtype" keyword
np.array([1,2,3,4.0], dtype = 'int')
```

```python
#you can create multidimensional numpy arrays
np.array([range(i,i+3) for i in[2,5,6]])
```

```python
#more verbose way of doing the above
a=[]
for i in [2,5,6]:
    a.append((range(i,i+3)))
np.array(a)
```

```python
#numpy as built in functions for creating arrays from scratch
#create a length 10 array with floats
np.zeros(10,dtype = 'float')
```

```python
#create multi-dimensional array filled with ones
np.ones((3,5),dtype='float')
```

```python
#create a multi-dimensional array with specific values
np.full((3,5),80085)
```

```python
#create an array filled with a linear sequence. 
#linear sequence is between 0 and 20 and increments at step levels = 2
np.arange(0,20,2)
```
