```python
import pandas as pd
import os
import seaborn as sns
import plotly.offline as plt
```

```python
#ZIP CODE AND MHA DATA FILE
sorted_zipmha20 = pd.read_csv("../datasets/bah_cost_of_living/sorted_zipmha20.txt",
                              delimiter = " ",
                              names = ['ZIPCODE','MHA'])

#Military Housing Areas (MHA) DATA FILE
mhanames20 = pd.read_csv("../datasets/bah_cost_of_living/mhanames20.txt",
                        delimiter = ";",
                        names = ['MHA','NAME_DESCRIPTION'])

#without dependents
bahwo20 = pd.read_csv("../datasets/bah_cost_of_living/bahwo20.txt",
                     header=None)
bahwo20.columns = ['MHA','E1','E2','E3','E4','E5','E6','E7','E8','E9','W1','W2','W3','W4','W5','O1E','O2E','O3E','O1','O2','O3','O4','O5','O6','O7','O8','O9','O10']

#with dependents
bahw20 = pd.read_csv("../datasets/bah_cost_of_living/bahw20.txt",
                    header = None)
bahw20.columns = ['MHA','E1','E2','E3','E4','E5','E6','E7','E8','E9','W1','W2','W3','W4','W5','O1E','O2E','O3E','O1','O2','O3','O4','O5','O6','O7','O8','O9','O10']

#zip to FIPS
ZIP_COUNTY_FIPS = pd.read_csv("../datasets/ZIP-COUNTY-FIPS_2017-06.csv")
```

```python
zip_bah = bahwo20.merge(sorted_zipmha20,
             left_on = 'MHA',
             right_on = 'MHA')

zip_bah_e5=zip_bah[['MHA','ZIPCODE','E5']]

zip_bah_e5_FIPS = zip_bah_e5.merge(ZIP_COUNTY_FIPS,
                                  left_on='ZIPCODE',
                                  right_on='ZIP')

zip_bah_e5_FIPS = zip_bah_e5_FIPS[['STCOUNTYFP','E5']]

zip_bah_e5_FIPS.groupby('STCOUNTYFP').mean().reset_index(inplace=True)
```

```python
import plotly.figure_factory as ff

import numpy as np
import pandas as pd

df_sample = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/minoritymajority.csv')
df_sample_r = df_sample[df_sample['STNAME'] == 'California']

values = df_sample_r['TOT_POP'].tolist()
fips = df_sample_r['FIPS'].tolist()

colorscale = [
    'rgb(193, 193, 193)',
    'rgb(239,239,239)',
    'rgb(195, 196, 222)',
    'rgb(144,148,194)',
    'rgb(101,104,168)',
    'rgb(65, 53, 132)'
]

fig = ff.create_choropleth(
    fips=fips, values=values, scope=['CA', 'AZ', 'Nevada', 'Oregon', ' Idaho'],
    binning_endpoints=[14348, 63983, 134827, 426762, 2081313], colorscale=colorscale,
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5}, round_legend_values=True,
    legend_title='Population by County', title='California and Nearby States'
)
fig.layout.template = None
fig.show()
```
