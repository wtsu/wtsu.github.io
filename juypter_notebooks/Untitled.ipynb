```python
import pandas as pd
import os
import seaborn as sns

#loading the data file
current_file_path = os.getcwd()
branch_file_path = os.path.dirname(current_file_path)
dataset_file_path = branch_file_path + '\\datasets'
data_file_path = dataset_file_path + '\\NASA_Labs_Facilities.xlsx'
data = pd.read_excel(data_file_path, header=1)

#cleaning up the data file's ambiguous column names
data.rename(columns ={"Unnamed: 1": "Department"}, inplace = True)
data.head()
```

```python
active_stations = data[data['Status'] == 'Active']
active_stations = active_stations[['State','Department','Status','Occupied']]

sns.violinplot(x=active_stations['State'], y=active_stations["Occupied"])
```
